```python
#Convert date and time to datetime and extract date, time, month, hours and day
merged_data['date'] = pd.to_datetime(merged_data['date'])
merged_data['time'] = pd.to_datetime(merged_data['time'])
merged_data['month'] = merged_data['date'].dt.month
merged_data['hour'] = merged_data['time'].dt.hour
merged_data['monthname'] = merged_data['date'].dt.month_name().str[:3]
merged_data['day'] = merged_data['date'].dt.day_name().str[:3]

merged_data
```

## üí° QUESTION 9

```python
#Group revenue by month
revenue_by_month_name = merged_data.groupby('monthname')['revenue'].sum()
revenue_by_month_name
```

```python
revenue_by_month_name.idxmax()
```

```python
#creating a bar plot to group for the sales made each month

x = revenue_by_month_name.index
y = revenue_by_month_name.values
plt.bar(x,y,) 
plt.title('Revenue By Month')
plt.xlabel('Month')
plt.ylabel('Revenue')

#Saving the bar plot as a png file
plt.savefig('revenue_by_month_bar_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#creating a line plot to compare the sales made each month

plt.plot(revenue_by_month_name.index, revenue_by_month_name.values, color= 'blue', marker= 'o')
plt.title('Revenue By Month')
plt.xlabel('Month')
plt.ylabel('Revenue')

plt.savefig('revenue_by_month_line_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

#### ‚ú®- From the 2 plots shown above, it is evident that the month of July has the most sales

## üí° QUESTION 6

```python
#Group the dataframe by hour and revenue to find how much sales is made each hour
revenue_by_hour = merged_data.groupby('hour')['revenue'].sum().reset_index()

revenue_by_hour
```

```python
#Find the peak hour of sales
peak_hour = merged_data.groupby('hour')['revenue'].sum().idxmax()

peak_hour
```

```python
#Find the hour where low sales are made
low_sales_hour = merged_data.groupby('hour')['revenue'].sum().idxmin()

low_sales_hour
```

```python
#Create a line plot to show the peak hour of sales

plt.plot(revenue_by_hour['hour'], revenue_by_hour['revenue'], color= 'blue', marker= 'o')
plt.title('Sales Made Per Hour')
plt.xlabel('Hour')
plt.ylabel('Revenue')

#Save the line plot as a PDF file
plt.savefig('revenue_by_hour_line_plot.pdf', dpi=300, bbox_inches='tight')
#Save the line plot as a png file
plt.savefig('revenue_by_hour_line_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

##### ‚ú® From the graph above, it shows that the peak hour of sales is 12, and the least is 9.
